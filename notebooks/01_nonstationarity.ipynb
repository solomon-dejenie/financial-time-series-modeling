```python
import os
import numpy as np
import pandas as pd

from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.vector_ar.vecm import coint_johansen

# -----------------------------
# Create project folders
# -----------------------------
os.makedirs("data/raw", exist_ok=True)

# -----------------------------
# Generate sample macro data
# -----------------------------
np.random.seed(42)
n = 50

df = pd.DataFrame({
    "GDP": np.cumsum(np.random.normal(0.1, 0.2, n)),
    "PCE": np.cumsum(np.random.normal(0.08, 0.15, n)),
    "GPDI": np.cumsum(np.random.normal(0.09, 0.18, n)),
    "GCE": np.cumsum(np.random.normal(0.07, 0.16, n))
})

# Save and reload data
df.to_csv("data/raw/macro_data.csv", index=False)

data = pd.read_csv("data/raw/macro_data.csv")

print(data.head())


# -----------------------------
# ADF Unit Root Test
# -----------------------------
adf_result = adfuller(data["GDP"])

print("\nADF Statistic:", adf_result[0])
print("p-value:", adf_result[1])


# -----------------------------
# Johansen Cointegration Test
# -----------------------------
coint_data = data[["GDP", "PCE"]]

jres = coint_johansen(
    coint_data,
    det_order=0,
    k_ar_diff=1
)

print("\nTrace Statistics:")
print(jres.lr1)
```
